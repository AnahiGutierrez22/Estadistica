##Ejercicio 1
Una revision de los pesos de 500 estudiantes del ITD, evidenci que la media es de 70 kilogramos y la desviacin estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:
{x}

Generar aleatoriamente las edades de los 500 estudiantes con una media de 70 kg, y una desviacin estandar de 4 kg.
 Realizar un histogrma de las edades
Graficar la funcin de densidad de probabilidad (PDF).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribucion normal
media = 70 # Media
desv_std = 4 #Desviacin estndar
num_datos = 500
#generamos los datos de manera aleatoria y lon ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos [0:50] # Mostrar los primeros 50
```

```python
#Crear un rango de valores para crera una densidad
x = np.linspace(media - 3*desv_std, media + 3*desv_std, num_datos)

#Calcular la pdf de la distribucion normal

pdf = 1/ (desv_std * np.sqrt(2* np.pi))* np.exp(-(x- media)**2/(2* desv_std**2))
```

```python
#Imprimir solo 50 valores
print ("Imprimir los primeros 50 valores de x ", x [0:50], " , y los primeros 50 pdf ", pdf [0:50])
pdf[0:50]
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
pdf.size, num_datos
```

```python
#Visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label= 'Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title(' Funcion de densidad de probabilidad (PDF) - Distribucin Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.legend()
plt.grid(True)
plt.show()
```

##EJERCICIO 2
Una revision de los pesos de 500 estudiantes del ITD, evidenci que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

Cuantos estudiantes pesan menos de 75 kg.

Cuantos estudiantes pesan menos de 62 kg.

Cuantos estudiantes pesan mas de 79 kg.

Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funcin de distribucin acumulativa (CDF) de la distribucin normal

```python
plt.figure(figsize=(6,4))

#parametros de la distribucion normal
media = 70 #media
desv_std = 4 #desviacion estandar
total_estudiantes = 500

#VARIABLE para Calculo de la probabilidad de queun estudiante pese menos de 75 kg
peso_limite = 79

#primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)




```

```python
#Calculo del numero de estudiantes que oesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("La probabilidad de que un estudiante pese menos de 75kg es de: ", prob_menos_75kg)
print("El numero de estudiantes que pesan menos de 75kg es de: ", estudiantes_menos_75kg)


 #Graficar la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4*desv_std, 1000)
y= norm.pdf(x, media, desv_std)

plt.plot(x,y, 'b', linewidth=2)
plt.fill_between(x,y, where=(x <= peso_limite), color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese menos de 62kg, \n $\mu= 70$, $\sigma=4$, X=500')
plt.xlabel('PesO (KG)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

##Cuantos estudiantes pesan menos de 62 kg

```python
plt.figure(figsize=(6,4))

#Parmetros de La distribucin normal
media = 70 # Media
desv_std= 4 # Desviacin estndar
total_estudiantes= 500
#VARIABLE para Clculo de La probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 62

#Primero necesitamos calcular La probabilidad de que un estudiante pese menos
# de 75 kg utilizando la funcin de distribucin acumulativa (CDF) de la distribucin normal
prob_menos_62kg =norm.cdf(peso_limite, media, desv_std)

#Clculo del nmero de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg= total_estudiantes * prob_menos_62kg
print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Nmero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_62kg)
```

```python
# Calcular la distribucin normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

#Graficar la distribucin normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

CUANTOS ESTUDIANTE PESAN MAS DE 79 KG

```python
plt.figure(figsize=(6,4))

#parametros de la distribucion normal
media = 70 #media
desv_std = 4 #desviacion estandar
peso_limite = 79
total_estudiantes = 500

#calculamos la probabilida de que un estudiante pese menos o igual a 79kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

#la probabilidad de que un estudiante pese mas de 79kg
prob_mas_79kg = 1 - prob_menos_79kg

#Calculamos el numero aproximado de estudiantes que pesan mas de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("La probabilidad de que un estudiante pese mas de 79kg es de: ", prob_mas_79kg)
print("El numero aproximado de estudiantes que pesen mas de 79kg es de: ", estudiantes_mas_79kg)






```

```python

#Calcular la funcion de desnsidad de probabilidad de La distribucion normal
x = np.linspace(media - 4*desv_std, media + 4*desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x,y, 'b', linewidth=2)
plt.fill_between(x,y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabiliadd de que un estudaiante pese mas de 79kg  \n $\mu= 70$, $\sigma=4$, X=500')

plt.xlabel('PesO (KG)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
