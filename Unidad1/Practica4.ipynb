```python
import random
```

```python
# lista, vector que representa la edad de 100 personas
edades =[]
# ciclo  para generar numeros aleatorios
for i in range (100):
  #generar el numero aleatorio con la funcion randit y se guarda en la variable numero_aleatorio
  numero_aleatorio= random.randint(18, 70)
# Agreganos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)
print (edades)
```

```python
#libreria para datos
import pandas as pd
```

```python
df= pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# calcular la tabla de distribucion de frecuencias

tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias  = tabla_frecuencias.sort_values(by='Edad')

# imprimir tabla de distribucion
print("Tabla de distribucin de frecuencias: ")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa']= tabla_frecuencias ['Frecuencia']/ 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']= tabla_frecuencias['Frecuencia']/100 * 100
tabla_frecuencias
```

```python
# cumsum para calcular la suma acumulada
tabla_frecuencias['Frec Acumulada']= tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

```python
# Regla de Sturges
#  es una regla que sirve para calcular el numero de clases o intervalos idondeo en los que se debe dividir un conjunto de datos
# establece un numero de clases es igual a uno mas el logaritmo en base des del numero drotal de datos
# c = 1 + log2 (N)
# Donde N es el tamao de la muestra
```

```python
import math
```

```python
#  volvemos a usar la lkista de edades
print (edades)
```

```python
# crear un data frame con las edades
df = pd.DataFrame ({'Edad': edades})
df
```

```python
#  calcular el nuemro d eintervalos utilizando la regla de sturges
n = len (edades)
k= int (1+ math.log2(100))
print ( " Valor de n = ", n , "\n Nmero de intervalos = ", k)
```

```python
# crear la tabla de frecuencias
tabla_frecuencias = pd.cut (df['Edad'], bins = k, include_lowest=True ). value_counts().reset_index()
tabla_frecuencias.columns= ['Clase', 'Frec']
tabla_frecuencias
```

```python
#  ordenar la tabla en base a las clases
# sort_values es el que ordena
tabla_frecuencias = tabla_frecuencias.sort_values (by= 'Clase')
tabla_frecuencias
```

```python
# generar columnas siguientes
tabla_frecuencias['Frec Rel']= tabla_frecuencias ['Frec']/ n
tabla_frecuencias['Frec Porc']= tabla_frecuencias ['Frec']/ n * 100
tabla_frecuencias['Frec Ac']= tabla_frecuencias ['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac']= tabla_frecuencias ['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac']= tabla_frecuencias ['Frec Porc'].cumsum()
#  Mostrar la atbal de frecuencias
print (tabla_frecuencias)
```
