```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
autos2=pd.read_csv('/content/autos_corregidos.csv')
autos2.head()
```

##Regresion lineal simple

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
```

```python
lm = LinearRegression()
#Variable independientes o variables predictorias
x = autos2 [["city-mpg"]]

#variable dependiente o variable objetiva a predecir
y = autos2 [["price"]]

x.head(10)
```

```python
#Entrenar el algoritmo
lm.fit(x,y)
```

```python
#realizamos la fase de prediccion
Ypredict = lm.predict(x)
```

```python
#Valor de la pendiente
lm.intercept_
```

```python
#Valor del coeficiente o peso de la variable
lm.coef_
```

```python
lm.score(x,y)
```

```python
from sklearn.model_selection import train_test_split
#Dividir el dataset en dos partes, una para entrenamiento de datos y otra para probar
autos2_train, autos2_test = train_test_split(autos2, random_state= 0)
```

##Regresion lineal multiple

```python
autos2[["horsepower", "engine-size", "city-mpg", "wheel-base", "curb-weight", "price" ]].corr()
```

```python
x_autos2_train = autos2_train[["horsepower", "engine-size", "city-mpg", "wheel-base", "curb-weight", "bore" ]]
# variables predictoreas para la fase de entrenamiento

y_autos2_train = autos2_train[["price"]]
# variable a predecir para la fase de entrenamiento
```

```python
x_autos2_test = autos2_test[["horsepower", "engine-size", "city-mpg", "wheel-base", "curb-weight", "bore" ]]
# variables predictoreas para la fase de prueba

y_autos2_test = autos2_test[["price"]]
# variable a predecir para la fase de prueba
```

```python
lm.fit(x_autos2_train, y_autos2_train)
#fase de entrenamiento del modelo lineal lm METODO FIT
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_autos2_train, y_autos2_train)
```

Fase de prueba

```python
#fase de prueba del modelo lineallm usando el metodo PREDICT
yhat= lm.predict(x_autos2_test)
```

```python
lm.score(x_autos2_test, y_autos2_test)
```

```python
import seaborn as sns
plt.figure(figsize= (6,4))

ax1 = sns.distplot(y_autos2_test["price"], hist=False, color="r")
ax1= sns.distplot(yhat, hist=False, color="b", ax=ax1)

plt.title("Precio Real vs Precio Prediccion")
ax1.legend(labels=["Valor Real", "Valor Prediccion"])

plt.show()
plt.close()
```

```python
plt.figure(figsize= (6,4))

ax1 = sns.kdeplot(y_autos2_test["price"],multiple="stack",alpha=0.5, color="r")
ax1= sns.kdeplot(pd.Series(yhat.flatten()), multiple="stack",alpha=0.5,color="r")

plt.title("Precio Real vs Precio Prediccion")
ax1.legend(labels=["Valor Real", "Valor Prediccion"])

plt.show()
plt.close()
```
