##Grupo 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statistics import *
```

```python
grupo1=pd.read_csv('/content/grupo1.csv')
grupo1.head()
```

```python
#Estadistica descriptiva
grupo1.describe()
```

```python
grupo1.info()
```

```python
# Histograma
plt.figure(figsize=(6,4))
sns.histplot(grupo1["Calificacion"], bins=10, kde=True, color='blue')
plt.title("Grupo 1 - Histograma de Calificaciones")
plt.xlabel("Calificacion")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
```

```python
lm = LinearRegression()
x = grupo1 [["HorasEstudio"]]
y = grupo1 [["Calificacion"]]

x.head(10)
```

```python
#Entrenar el algoritmo. Entrenar un modelo de regresin lineal simple usando scikit-learn o statsmodels
lm.fit(x,y)
```

```python
Ypredict = lm.predict(x)
```

```python
#Obtener la ecuacin de la lnea de regresin (pendiente e intercepto). Valor de la pendiente
lm.intercept_
```

```python
#Valor del coeficiente de correlacion de pearson entre las horas de estudio y las calificaciones
lm.coef_
```

```python
lm.score(x,y)
```

```python
#Graficar los datos junto con la lnea de regresin.
#Incluir un grfico de dispersin (scatter plot) y la lnea de regresin.

import seaborn as sns
import matplotlib.pyplot as plt
#definir tamao
plt.figure(figsize= (6,4))
#Crear el diagrama de dispersion con lineas de regresion
sns.set(style= "whitegrid")
sns.regplot(x=x, y=y, scatter=True,  color="blue")
plt.title('Diagrama de dispersin con lnea de regresin- Grupo 1 ')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```

1. Explicar cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?

En este primer grupo se observa que esta muy disperso, es como si las horas de estudio no afectaran micho, ya que hat pocas horas que tienen una alta calificacion y otras que tienen poca calificacion, de la misma manera con mas horas de estudio esta muy variado
2. Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y
direccin de la relacin entre las variables?

Nos dio que es una direccion negativa, como se mencionaba arriba no va en aumento por mas horas estudiadas

3. El modelo de regresin lineal es estadsticamente significativo?

No es tan significativo, para ser significativo deberia de ser un valor mas alto

4. Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?

Este modelo no creo que sea muy confiable porque no tiene un aumento fijo por las horas estudiadas, esta muy variado por lo que pienso que no es confiable.
